```python
import cv2 
import matplotlib.pyplot as plt
import numpy as np
```

```python
logo = np.ones((600, 1000, 3), dtype = "uint8") * 127
plt.imshow(logo)
```

```python

logo[210:310, 210:310] = (246, 83, 20)
logo[320:420, 210:310] = (0, 161, 241) 
logo[210:310, 320:420] = (124, 187, 0) 
logo[320:420, 320:420] = (255, 187, 0) 
```

```python

cv2.putText(logo, text= "Microsoft", org=(450, 350), fontFace=cv2.FONT_HERSHEY_TRIPLEX, fontScale=3, color=(255, 255, 255),thickness=2)
plt.imshow(logo)
```
