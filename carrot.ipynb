```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img_BGR = cv2.imread("Carrot.jpg")
img_RGB = cv2.cvtColor(img_BGR, cv2.COLOR_BGR2RGB)

plt.imshow(img_RGB)
```

```python
rows, cols, _ = img_RGB.shape
R, G, B = cv2.split(img_RGB)
print('RED')
plt.imshow(R)
plt.show()

print('GREEN')
plt.imshow(G)
plt.show()

print('BLUE')
plt.imshow(B)
plt.show()
```

```python
R_new = G
G_new = G
B_new = B
```

```python
img_result = cv2.merge((R_new, G_new, B_new))
plt.imshow(img_result)
```
